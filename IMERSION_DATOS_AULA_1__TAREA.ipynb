```python
import pandas as pd

inmuebles = pd.read_csv("/content/drive/MyDrive/inmuebles_bogota (1).csv")
inmuebles
```

Promedio de área de todos los inmuebles en los barrios

```python
inmuebles[["Barrio","Área"]].groupby(["Barrio"]).mean().sort_values(by=["Área"], ascending=False)
```

Promedio de área de todos los inmuebles en los barrios. El Top 10

```python
inmuebles[["Barrio","Área"]].groupby(["Barrio"]).mean().sort_values(by=["Área"], ascending=False).head(10)
```

Grafico del promedio de área de todos los inmuebles en los barrios. El Top 10

```python
inmuebles[["Barrio","Área"]].groupby(["Barrio"]).mean().sort_values(by=["Área"], ascending=False).head(10).plot(kind='bar')
```
